Q5. Loyal customers have the lowest average spending power and new customers have the highest spending power.

```python
#Q6 create a new column based on average spending per order_id
ords_prods_merge['spending_power'] = ords_prods_merge.groupby(['order_id'])['prices'].transform(np.mean)
```

```python
#Deriving Columns with loc()
ords_prods_merge.loc[ords_prods_merge['spending_power'] < 10, 'Spender_flag'] = 'Low spender'
ords_prods_merge.loc[ords_prods_merge['spending_power'] >= 10, 'Spender_flag'] = 'High spender'
```

```python
ords_prods_merge['Spender_flag'].value_counts(dropna=False)
```

Most of the orders are low spender.

```python
#Q7 create a new column based on the medium of "days_since_prior_order"
ords_prods_merge['frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#Deriving Columns with loc()
ords_prods_merge.loc[ords_prods_merge['frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['frequency'] <= 20) & (ords_prods_merge['frequency'] > 10), 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['frequency'] <= 10, 'frequency_flag'] = 'frequent customer'
```

```python
ords_prods_merge['frequency_flag'].value_counts(dropna=False)
```

Majority of the customers are frequent customers.
